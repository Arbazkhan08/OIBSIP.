```python
p = df.groupby("Product Category")[["Quantity", "Total Amount"]].sum()
p
```

```python
p.plot(kind = "bar", figsize = (12, 5))
plt.title("Quantity v/s Total Amount")
plt.ylabel("Quantity and Total Amount")
plt.show()
```

```python
cate=df["Product Category"].value_counts()
cate
```

```python
explode_list=[0,0.1,0.1]
color_list = ["Orange", "Green", "Blue"]
cate.plot(kind="pie", figsize=(8,6))
plt.title("Product Category", fontsize=(14))
plt.axis("equal")
plt.show()
```

```python
# Assuming cate is your data series
cate = pd.Series([30, 40, 30], index=["Electronic", "clothing", "Beauty"])
explode_list = [0, 0.1, 0.1]
color_list = ["Orange", "Green", "Blue"]
plt.figure(figsize=(8, 6))
cate.plot(kind="pie", explode=explode_list, colors=color_list, autopct='%1.1f%%')

plt.title("Product Category", fontsize=14)
plt.axis("equal") 
plt.show()
```
